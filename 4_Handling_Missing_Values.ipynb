# dropna & fillna

```python
import pandas as pd

var = pd.read_csv("IMDB_Movies.csv")

var
```

# dropna()

```python
# Droping Null values

var.dropna()

# dropna() by defalut remove rows having null value
```

```python
# to delete column having null values

var.dropna(axis=1)
```

```python
var
```

```python
# to delete rows which have all teh cellls of null values i.e. whole row is null

var.dropna(how="all")
```

```python
var
```

```python
# Delete rows having null values for certain attribute/column

var.dropna(subset=["actor_3_facebook_likes"])
```

```python
# Remove the null values from the actual dataframe
var.dropna(inplace=True)

var
```

```python
var
```

# fillna()

```python

var1 = pd.read_csv("IMDB_Movies.csv")

var1
```

```python
# replace all null values with certain value

var1.fillna("python")
```

```python
var1
```

```python
# fill perticular columns null values with perticular data

var1.fillna({"actor_3_facebook_likes":0, "gross":0, "language":"Marathi", "country":"India", "title_year":2023})
```

```python
# fill null value with previous data of that value

# forward filling (fills null value with the previous value)
var1.fillna(method="bfill")
```

```python
# fill null value with forward data of that value

# forward filling (fills null value with the next value)

var1.fillna(method="ffill")
```

```python
# We can also make them will along the axis i.e. neihboureing column values

var1.fillna(method="ffill", axis=1)
```

```python
var1.fillna(0,inplace=True)
var1
```

```python
# making original var for editing
var1 = pd.read_csv("IMDB_Movies.csv")

var1

# filling the first N number of null values using limit parameter
var1.fillna("python", limit = 3)
```

# Replace and Interpolate function

```python
var3 = pd.read_csv("IMDB_Movies.csv")

var3
```

```python
# To replace all some value in the whole dataframe

var3.replace(to_replace=0, value=111)
```

```python
# Replace many values using the one value

var3.replace([1,2,3,4,5,6,7,8],22)
```

```python
# Replace using the regular expression

var3.replace("[A-Za-z]", "py", regex=True)
```

```python
var3
```

```python
# Replace data of perticular column using regular expression

var3.replace({"director_name":"[A-Z]"},22,regex=True)
```

```python
# replace using the method 

var3.replace(1,method="ffill",limit=2)
```

```python
# What does interpolate do in Python?
# Interpolation in Python is a technique used to estimate unknown data points between two known data points. 
# In Python, Interpolation is a technique mostly used to impute missing values in the data frame or series while preprocessing data
```

```python
var3.interpolate()

# it only fills number data not charts/strings
# there are alot of methods to use while filling those missing data defult parameter for method is linear
```

```python
var3.interpolate(method="linear",asix=0)
```

```python
# in interpolate there is a limit_area parameter which gives when you should perform interpolate replace.

var3.interpolate(limit_area= "outside")
```

# Merge

```python
import pandas as pd

var1 = pd.DataFrame({"S":[1,2,3,4], "B":[11,22,33,44]})
var2 = pd.DataFrame({"S":[1,2,3,4], "C":[23,22,33,44]})

pd.merge(var1, var2, on="S")
```

```python
# it only give the merge on the common data

var1 = pd.DataFrame({"S":[1,2,3,4], "B":[11,22,33,44]})
var2 = pd.DataFrame({"S":[1,2,3,5], "C":[23,22,33,44]})

pd.merge(var1, var2, on="S")
```

```python
# inner, left, right methods using the how parameter

var1 = pd.DataFrame({"S":[1,2,3,4], "B":[11,22,33,44]})
var2 = pd.DataFrame({"S":[1,2,3,5], "C":[23,22,33,44]})

pd.merge(var1, var2, how= "inner")
```

```python
var1 = pd.DataFrame({"S":[1,2,3,4], "B":[11,22,33,44]})
var2 = pd.DataFrame({"S":[1,2,3,5], "C":[23,22,33,44]})

pd.merge(var1, var2, how= "left",)
```

```python
var1 = pd.DataFrame({"S":[1,2,3,4], "B":[11,22,33,44]})
var2 = pd.DataFrame({"S":[1,2,3,5], "C":[23,22,33,44]})

pd.merge(var1, var2, how= "right", left_index = True, right_index = True,  suffixes = ("name","python"))
```

```python
var1 = pd.DataFrame({"S":[1,2,3,4], "B":[11,22,33,44]})
var2 = pd.DataFrame({"S":[1,2,3,5], "C":[23,22,33,44]})

pd.merge(var1, var2, how= "outer", indicator = True)
```

# Concat

```python
# lets create two series and concat them

sr1 = pd.Series([1,2,3,4])
sr2 = pd.Series([11,22,32,42])

pd.concat([sr1,sr2])
```

```python
d1 = pd.DataFrame({"S":[1,2,3,4], "B":[11,22,33,44]})
d2 = pd.DataFrame({"S":[1,2,3,5], "C":[23,22,33,44]})

pd.concat([d1,d2])
```

```python
# Handle this missing data using the join Parameter

d1 = pd.DataFrame({"S":[1,2,3,4], "B":[11,22,33,44]})
d2 = pd.DataFrame({"S":[1,2,3,5], "C":[23,22,33,44]})

pd.concat([d1,d2], join="inner")
```

```python
# Creating the key of the concat of the datafrmes

d1 = pd.DataFrame({"S":[1,2,3,4], "B":[11,22,33,44]})
d2 = pd.DataFrame({"S":[1,2,3,5], "C":[23,22,33,44]})

pd.concat([d1,d2], join="outer", keys = ["d1", "d2"])
```

```python
# Concat along column axis

d1 = pd.DataFrame({"S":[1,2,3,4], "B":[11,22,33,44]})
d2 = pd.DataFrame({"S":[1,2,3,5], "B":[23,22,33,44]})

pd.concat([d1,d2], axis =1)
```

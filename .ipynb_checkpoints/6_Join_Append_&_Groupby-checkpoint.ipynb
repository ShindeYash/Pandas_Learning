# Join


```python
import pandas as pd

var1 = pd.DataFrame({"A":[1,2,3,4], "B":[1,22,33,444]})
var2 = pd.DataFrame({"C":[10,20,30], "D":[11,2,3]})

var1.join(var2)
```

```python
var1 = pd.DataFrame({"A":[1,2,3,4], "B":[1,22,33,444]}, index=["a","b","c","d"])
var2 = pd.DataFrame({"C":[10,20], "D":[11,2]}, index= ["a", "b"])

var2.join(var1)
```

```python
var1 = pd.DataFrame({"A":[1,2,3,4], "B":[1,22,33,444]}, index=["a","b","c","d"])
var2 = pd.DataFrame({"C":[10,20], "D":[11,2]}, index= ["a", "b"])

var2.join(var1, how = "right")
```

```python
# Outer works as Union and Inner works as Inersection

var1 = pd.DataFrame({"A":[1,2,3,4], "B":[1,22,33,444]}, index=["a","b","c","d"])
var2 = pd.DataFrame({"C":[10,20], "D":[11,2]}, index= ["a", "b"])

var2.join(var1, how = "outer")
```

```python
# to handle the nameerror of common name for columns
# we can use lsuffix to change the column names

var1 = pd.DataFrame({"A":[1,2,3,4], "B":[1,22,33,444]}, index=["a","b","c","d"])
var2 = pd.DataFrame({"C":[10,20], "B":[11,2]}, index= ["a", "b"])

var2.join(var1, how= "outer", lsuffix="_to", rsuffix="_me")
```

# Append

```python
# Used to add two dataframes
# Major difference between join an append function
# Column values same error was in the join function but there is no such error for append function
```

```python
var1 = pd.DataFrame({"A":[1,2,3,4], "B":[1,22,33,444]}, index=["a","b","c","d"])
var2 = pd.DataFrame({"C":[10,20], "B":[11,2]}, index= ["a", "b"])

var1.append(var2)
```

```python
# to ignore index

var1.append(var2, ignore_index=True)
```

# Groupby

```python
var = pd.DataFrame({"name": ["a","b","c","d","a","b","a","b","c","c","d","a"], "sno1": [23,34,55,55,33,45,34,56,35,24,53,56]})

var
```

```python
# here we can group data by aa bb cc such way using groupby function

var_new = var.groupby("name")

var_new
```

```python
# how to show that groupby output

for x,y in var_new:
    print(x)
    print(y)
    print()
```

```python
var_new.get_group("d")
```

```python
var_new.min()
```

```python
var_new.max()
```

```python
var_new.mean()
```

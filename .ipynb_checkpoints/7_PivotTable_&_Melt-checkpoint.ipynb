# Melt()

```python
import pandas as pd

var = pd.DataFrame({"days": [1,2,3,4,5], "eng":[10,12,14,16,18], "maths": [17,18,19,14,16]})
var
```

```python
pd.melt(var, id_vars=["eng"], var_name="python", value_name="yash")
```


# Pivot()

```python
var = pd.DataFrame({"days": [1,2,3,4,5], "stu":list("abcbc") ,"eng":[10,12,14,16,18], "maths": [17,18,19,14,16]})
var
```

```python
var.pivot(index="days", columns="stu")
```

```python
var.pivot(index="days", columns="stu", values="eng")
```

```python
# here we can work with this using sum or average finding on such data
```

```python
var.pivot_table(index = "stu", columns="days",
               aggfunc="sum")
```

```python
var.pivot_table(index = "stu", columns="days",
               aggfunc="sum", margins=True)
```

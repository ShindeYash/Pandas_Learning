```python
# How to use loc for getting complete pertiular columns

csv_1.loc[:,["Column1", "location"]]
```

```python
# How to use loc for getting complete pertiular rows

csv_1.loc[[2,3],:]
```

```python
# How to use loc for getting pertiular columns and rows

csv_1.loc[[2,3],["Column1", "location"]]
```

### Using iloc fucntion

```python
help(csv_1.iloc)
```

```python
print(csv_1.iloc[0,2])

print(csv_1.iloc[0,7])
```

```python
# get columns using iloc function

csv_1.iloc[:,[2,3]]
```

```python
# get columns using iloc function

csv_1.iloc[[2,3],:]
```

```python
# Difference Between loc and iloc is that
# in iloc you can access the columns using the index number of the columns (0, 1,2,3 etc)
# in loc you can't access the columsn using the index number of the columns (0, 1,2,3, etc) you have to use the actual name of the column
```

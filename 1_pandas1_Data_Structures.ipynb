# Series

```python
import pandas as pd

x = [4,5,6,7,8,4,"yash",5]

var = pd.Series(x)

print(var)
print(type(var))

print(var[3])
```

Change Index of Series

```python
x = [4,5,6,7,8,4,"yash",5]

var1 = pd.Series(x, index=['a','b','c','d','e','f','g','h'])

print(var1)
```

Datatype Change in Series

```python
x = [4,5,6,7,8,4,5]

var = pd.Series(x, dtype="float", index=['a','b','c','d','e','f','g'])

print(var)
print(type(var))
```

Nameing of Data

```python
x = [4,5,6,7,8,4,5]

var3 = pd.Series(x, dtype="float", index=['a','b','c','d','e','f','g'], name = "yash_numbers")

print(var3)
```

```python
dic = {"name":["python", 'c', 'c++','java'], "pos":[12,13,14,15], "rank":[1,4,3,2]}

var3 = pd.Series(dic)

print(var3)
print(var3[2])
```

```python
s = pd.Series(12,index=[1,2,3,4,5])


print(s)
print(type(s))
```

```python
s1 = pd.Series(12, index = [1,2,3,4,5,6])
s2 = pd.Series(12, index = [1,2,3,4])

print(s1+s2)
```

# DataFrame

```python
# To create DataFrame You need list or Dict

l = [1,2,3,4,5,6]

var = pd.DataFrame(l)

var
print(type(var))
print(var)
```

```python
d = {"a":[1,2,3,4,5], "s":[6,7,8,9,10], "d":[11,22,33,44,55], "e":['y','a','s','h','e']}

var1 = pd.DataFrame(d)

print(type(var1))
print(var1)
```

```python
d1 = {"a":[1,2,3,4,5,6], "s":[6,7,8,9,10,55], "d":[11,22,33,44,55,110], "e":['y','a','s','h','e','d']}

var2 = pd.DataFrame(d1)


print(type(var2))
print(var2)
```

```python
var3 = var1 + var2
print(var3)
```

Select Preticular Column

```python
var1 = pd.DataFrame(d, columns=['a','e','f'])

print(var1)
```

Change Index of DataFrame

```python
var1 = pd.DataFrame(d, columns=['a','e'], index=[1,2,3,4,5])
print(var1)
```

```python
print(var1['a'][4])
```

Create DataFrame Using Multidimenstional List

```python
list1 = [[1,2,3,4],list("jyash")]

var4 = pd.DataFrame(list1)

print(var4)
print(type(var4))
```

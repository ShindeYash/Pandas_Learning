# Arithmetic Operations

```python
import pandas as pd

var = pd.DataFrame({"A":[1,2,3,4], "b":[5,6,7,8]})

var
```

```python
var['c'] = var['A'] + var['b']

var
```

```python
var['c'] = var['A'] - var['b']

var
```

```python
var['c'] = var['A'] * var['b']

var
```

```python
var['c'] = var['A'] / var['b']

var
```

Logical Operations

```python
var1 = pd.DataFrame({"A":[10,20,30,40], "b":[15,16,17,18]})

var1
```

```python
var1["Python"] = var1["A"] <= 20
var1['Python1'] = var1['b'] >= 16

var1
```
